```python
stock_multi = twstock.realtime.get(['2330', '2454', '0050'])
stock_multi
```

```python
pd.DataFrame(stock_multi)
```

```python
df = stock_multi['2330']
df
```

```python
df['success']
```

```python
df = pd.DataFrame(df)
df
```

```python
stock_info = df.iloc[0:5, 1]
stock_realtime = df.iloc[5:, 2]
stock_data = stock_info._append(stock_realtime, ignore_index = True)
df['data'] = list(stock_data)
df = df.drop(['timestamp', 'info', 'realtime', 'success'], axis = 1)
df = df.T
df.columns = ['股票代碼', '股票代碼全', '股票名稱', '公司名', '下載時間', '最新成交量', '成交量', '累計成交量', '最佳5檔賣出價', '最佳5檔賣出量', '最佳5檔買進價', '最佳5檔買進量', '開盤價', '最高價', '最低價']
df
```

```python
stock_df = pd.DataFrame()

for stock in ['2330', '2454', '0050']:
    df_= pd.DataFrame(stock_multi[stock])
    stock_info = df_.iloc[0:5, 1]
    stock_realtime = df_.iloc[5:, 2]
    stock_data = stock_info._append(stock_realtime, ignore_index = True)
    df_['data'] = list(stock_data)
    df_ = df_.drop(['timestamp', 'info', 'realtime', 'success'], axis = 1)
    df_ = df_.T
    stock_df = stock_df._append(df_, ignore_index = True)
stock_df.columns = ['股票代碼', '股票代碼全', '股票名稱', '公司名', '下載時間', '最新成交量', '成交量', '累計成交量', '最佳5檔賣出價', '最佳5檔賣出量', '最佳5檔買進價', '最佳5檔買進量', '開盤價', '最高價', '最低價']
```
